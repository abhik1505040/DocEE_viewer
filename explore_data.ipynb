```python
import json
with open("DocEE-en.json") as f:
    json_data = json.load(f)

type_role_tuples = set([(k[2], l["type"]) for k in json_data for l in json.loads(k[3])])
```

```python
type2roles = dict()
for t1, t2 in type_role_tuples:
    type2roles[t1] = type2roles.get(t1, set()) | set([t2])
```

```python
max(len(k) for k in type2roles.values())
```

```python
color_scheme = [
    "#7aecec",
    "#bfeeb7",
    "#feca74",
    "#ff9561",
    "#999cff",
    "#c887fb",
    "#9cc9cc",
    "#ffeb80",
    "#ff8197",
    "#f0d0ff",
    "#00e1d9",
    "#e4e7d2",
    "#33ffd4",
    "#99ff33",
    "#ff9933",
    "#ffff33",
    "#ccffff",
    "#ff32ff",
    "#ffcc99",
    "#66ffcc",
    "#ffff66",
]

from IPython.display import Markdown, display

display(Markdown('<br>'.join(
    f'<span style="font-family: monospace">{color} <span style="color: {color}">████████</span></span>'
    for color in color_scheme
)))
```

```python
color_map = {}

for type, roles in type2roles.items():
    color_map.update(
        dict(zip([f"{type}█{k}" for k in roles], color_scheme))
    )

print(json.dumps(color_map, indent=4))
```
